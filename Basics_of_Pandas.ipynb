## Group by

```python
company=pd.read_csv('/content/Fortune 500 Companies US.csv', encoding='latin-1')
company.head(2)
```

```python
company.info()
```

```python
company['Revenues'] = company['Revenues'].apply(lambda x:x.replace(',','') and x.replace('$','')).astype('int')
company.head(2)
# company[company['Revenues'].str.contains('$30109')].shape
```

```python
companies = company.groupby('Company Name') #Not directly assessible
```

```python
len(companies)
```

```python
companies.size()
```

```python
companies.size().sort_values(ascending=False) #Start with the repeating companies
```

```python
companies.first() #DF for every group
```

```python
companies.last() #DF for every group
```

```python
companies.groups
```

```python
companies.get_group('3M') #Common
```

```python
companies.get_group('3M').shape
```

```python
companies['Revenues'].mean()
```

## IPL EDA

```python
delivery = pd.read_csv('/content/deliveries.csv', error_bad_lines=False)
```

```python
delivery.head(5)
```

```python
delivery.shape
```

```python
#Top 10 batsmen with the most runs in the 10 seasons of IPL
top_10_batsman = delivery.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
top_10_batsman.plot(kind='bar')
```

```python
#How many runs played by Virat Kohli in all seasons of IPL
delivery.groupby('batsman').get_group('V Kohli').shape
```
